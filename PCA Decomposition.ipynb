# PCA decomposition

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd
```

```python
ndims = 300
column_names = list(range(ndims))
df = pd.read_csv('docs_vector.csv', names = ['Index'] + column_names, index_col = 0, header=None)
df.head()
```

# PCA Decomposition

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
principalComponents = pca.fit_transform(df)
print('Variation Ratio: ', pca.explained_variance_ratio_)
principalDf = pd.DataFrame(data = principalComponents)
principalDf.index = df.index
principalDf.to_csv('docs_vector_reduced.csv')
principalDf.head()
```
